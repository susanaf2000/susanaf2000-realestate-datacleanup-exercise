#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
filtered_data = ds[ds['level5'].isin(populations)]

scaler = MinMaxScaler()


plt.figure(figsize=(12, 8))


for population in populations:
    pop_data = filtered_data[filtered_data['level5'] == population]
    prices = pop_data['price'].dropna().values.reshape(-1, 1)
    normalized_prices = scaler.fit_transform(prices)
    plt.hist(normalized_prices, bins=30, alpha=0.5, label=population)



plt.title('Normalized Price Distributions Across Populations')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

With this plot we can scale prices between 0 and 1 and have a better understanding of their price differences. The peaks indicating the prevalent price ranges and the overlaps indicating a common price range among these populations. We can therefore use this for a better understanding of markets trends withing these specific regions.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
filtered_data = ds[ds['level5'].isin(populations)]

getafe_data = filtered_data[(filtered_data['level5'] == 'Getafe') & filtered_data['price'].notna() & filtered_data['surface'].notna()]
alcorcon_data = filtered_data[(filtered_data['level5'] == 'Alcorcón') & filtered_data['price'].notna() & filtered_data['surface'].notna()]

getafe_data = getafe_data[getafe_data['surface'] > 0]
alcorcon_data = alcorcon_data[alcorcon_data['surface'] > 0]

getafe_data['price_per_m2'] = getafe_data['price'] / getafe_data['surface']
alcorcon_data['price_per_m2'] = alcorcon_data['price'] / alcorcon_data['surface']

plt.figure(figsize=(12, 8))

plt.hist(getafe_data['price_per_m2'], bins=30, alpha=0.5, label='Getafe', edgecolor='black')
plt.hist(alcorcon_data['price_per_m2'], bins=30, alpha=0.5, label='Alcorcón', edgecolor='black')

plt.title('Distribution of Price per Square Meter in Getafe and Alcorcón')
plt.xlabel('Price per square meter (€)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.show()
```

We can see in this plot that Getafe shows a wider range of prices per square meter. It suggest a more diverse property market, with lower and higher priced properties. In Alcorón, we can constate a distribution concentrated around the median, meaning we have a more uniform pricing structure across these properties. Alcorón in general has higher prices per square than Getafe

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import matplotlib.pyplot as plt
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
filtered_data = ds[ds['level5'].isin(populations)]

plot_data = filtered_data[['level5', 'price', 'surface']].dropna()

plt.figure(figsize=(12, 8))

for population in populations:
    subset = plot_data[plot_data['level5'] == population]
    plt.scatter(subset['surface'], subset['price'], label=population)

plt.xlabel('Surface Area')
plt.ylabel('Price')
plt.title('Comparison of Real Estate Prices and Surface Area in Different Populations')
plt.legend()
plt.show()
```
